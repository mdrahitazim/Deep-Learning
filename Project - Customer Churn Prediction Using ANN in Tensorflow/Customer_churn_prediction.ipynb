**One hot encoding**

there are many object datatype so need to convert to integer

```python
df1.dtypes
```

```python
for column in df:
    print(column)
```

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes == 'object':
            print(f'{column} : {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No', inplace = True)
```

```python
for column in df1:
    print(f'{column} : {df1[column].unique()}')
```

```python
col_to_replace = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
df1[col_to_replace] = df1[col_to_replace].replace({'Yes':1,'No':0})
```

```python
for column in df1:
    print(f'{column} : {df1[column].unique()}')
```

```python
#convert gender to 0 and 1
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
for column in df1:
    print(f'{column} : {df1[column].unique()}')
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
col_to_scale = ['tenure','MonthlyCharges','TotalCharges']
df1[col_to_scale] = scaler.fit_transform(df1[col_to_scale])
```

```python
for column in df1:
    print(f'{column} : {df1[column].unique()}')
```

```python
col_to_labelEncode = ['InternetService','Contract','PaymentMethod']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in col_to_labelEncode:
    df1[column] = le.fit_transform(df1[column])
```

```python
for column in df1:
    print(f'{column} : {df1[column].unique()}')
```

```python
df1.dtypes
```

**Train test split**

```python
from sklearn.model_selection import train_test_split
X = df1.drop('Churn',axis='columns')
y = df1['Churn']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

**Making a ANN model in Tensorflow using keras api**

```python
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(26,input_dim=19,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```
